## 3.1.Time Window (light logic)
---

```python
# -- 90일 단위의 윈도우를 형성해서 쪼갬
from datetime import datetime, timedelta

def windows90(df, start, window=None):
    start = datetime.strptime(start, "%Y-%m-%d")
    # end = datetime.strptime(end, "%Y-%m-%d")
    if window is None:
        window=[date.strftime("%Y-%m-%d") for date in pd.date_range(start=start, periods=90)]
    else:
        window=window[1:]+[window[-1]+timedelta(days=1)]
    return df.loc[df["de_dt"].isin(window)]

windows90(df_pr, "2021-01-01")
```

```python
# -- 첫 90일 간 데이터를 활용하여 상품-고객에서 매출에 대한 pivot table

df_ppv = pd.pivot_table(data=windows90(df_pr, "2021-01-01"),
               values="buy_am",
               index="cust",
               columns="pd_c",
               aggfunc="sum",
               fill_value=0).reset_index(drop=False)
```
